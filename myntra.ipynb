```python
reviews = review_html.find_all(("div", {"class":"detailed-reviews-userReviewsContainer"}))
reviews
```

```python
# Assuming review_html is already populated with page content

# Find all the review containers
reviews = review_html.find_all("div", {"class": "detailed-reviews-userReviewsContainer"})

# Check if any reviews were found
if not reviews:
    print("No reviews found with the specified class.")
else:
    for idx, review in enumerate(reviews):
        # For each review, find the specific wrapper for the review text
        review_comment = review.find_all("div", {"class": "user-review-userReviewWrapper"})
        user_rating = review.find("span", {"class": "user-review-starRating user-review-fiveStars"}).get_text().strip()

        # Find the container for the user name and date
        user_info = review.find_all("div", {"class": "user-review-left"})
        
        # Print debug information to check what's happening
        if not review_comment:
            print(f"Review {idx}: Empty review found. Check the structure.")
        else:
            print(f"Rating {user_rating[idx].text}")
            if user_info:
                first_span = user_info[0].find("span")  # Extract the first span only
                if first_span:
                    print(f"User Name {idx}: {first_span.text.strip()}")
                else:
                    print(f"User Name {idx}: Not found")

            # Print the review comment
            for text in review_comment:
                print(f"Review {idx}: {text.text.strip()}")
```
